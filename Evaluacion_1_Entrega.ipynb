## 1° Ejercicio (0,5 puntos):

Resuelva el problema de obtener la sumatoria de los primeros n números consecutivos (1+2+...+n) usando listas y bucles

```python
from random import randint

def Sumatoria(n):
    sumatoria = 0
    lista_n = []
    for i in range(1,n+1):
        lista_n.append(i)
        sumatoria = sumatoria + i
    print(f"Lista Generada ---> {lista_n}")
    print(f"Sumatoria: {sumatoria}")
    return sumatoria

Sumatoria(randint(1,20))
```

## 2° Ejercicio (1 punto):

En Biwiser necesitan poder valorizar los proyectos que venden a sus clientes y también estimar la cantidad de analistas que requerirán para cumplir el proyecto en el plazo estipulado.
Para esto considere lo siguiente:
- Los analistas tiene máximo 40 horas semanales.
- Se le cobra al cliente 3 UF/hh consultoría
- El costo de cada analista es 2.5 UF/hh.

Debe crear un código que use como parámetro la cantidad de horas estimadas para desarrollar el proyecto y el plazo en semanas, ambos entregados __por el usuario__ y responda dentro de un mensaje con los cálculos solicitados.

__Ej :__ 

* Cantidad horas proyecto : 81 hh
* Semanas estimadas : 2 semanas

El output debería algo del estilo:
* Para realizar el proyecto necesita 2 analistas
* Cobro cliente:  243.0 UF
* El costo total de analistas es de: 202.5 UF
* El margen, considerando cobro clientes y costo de analistas es de:  40.5 UF


```python
from datetime import datetime
from dateutil.relativedelta import *

horas_estimadas = int(input("Ingrese las Horas Estimadas para el Desarrollo del Proyecto: "))
semanas_plazo = int(input("Ingrese el Plazo Requerido para el Proyecto:"))

def CalcularProyecto(horas,semanas):
    h_max_analista = 40
    valor_hora = 3
    costo_analista = 2.5
    hoy = datetime.today()
    semanas_plazo = relativedelta(weeks=semanas)

    total_analistas = horas / h_max_analista

    total_cobro_cliente = horas * valor_hora

    total_costo_analistas = horas * costo_analista

    margen = total_cobro_cliente - total_costo_analistas

    plazo_estimado = hoy + semanas_plazo
    
    # Outputs --->
    print("*"*22+" Detalles de Consultoría para Proyectos "+"*"*22)
    print(f">>> Para realizar el proyecto necesita {round(total_analistas)} analistas")
    print(f">>> Cobro Cliente: {total_cobro_cliente} UF")
    print(f">>> El costo total de analistas es de: {total_costo_analistas} UF")
    print(f">>> El margen, considerando cobro clientes y costo de analistas es de: {margen} UF")
    print("*"*84)
    print(f"Plazo Estimado de Entrega: {plazo_estimado.strftime('%d-%m-%Y')}")
    print("*"*84)

    
CalcularProyecto(horas_estimadas,semanas_plazo)
```

## 5° Ejercicio (0,5 puntos):

Escriba un programa que tome una lista de números como entrada y genere como output el producto de todos los números en la lista.

```python
from random import randint

def GeneradorNumeros(n):
    lista_num = []
    for i in range(1,n):
        i = randint(1,10)
        lista_num.append(i)
    return lista_num

def ProductoEnteros(numeros):
    producto = 1
    for numero in numeros:
        producto = producto * numero

    return producto

lista_numeros = GeneradorNumeros(randint(2,20))
producto = ProductoEnteros(lista_numeros)
print(f"Lista de {len(lista_numeros)} Números ---> {lista_numeros}")
print(f"El Producto de todos los Números es: {producto}")
```

## 6° Ejercicio (1 puntos):

Una cadena de salas de cine quiere captar más público familiar, por lo que hace tramos de precio según su edad:

| Rango Edad | Precio |
| :----: | :----: |
| Menos de 6 años | $0 |
| Entre 6 y 14 años | \$2.000 |
| Entre 15 y 17 años | \$4.000 |
| Mayor o igual a 18 años | \$5.000 |

Escriba una código en que el usuario ingrese su fecha de nacimiento y el computador responda con el precio de la entrada según su edad.

```python
from datetime import datetime
from dateutil.relativedelta import *

fechaActual = datetime.today()
fechaNacimiento = input("Ingrese su Fecha de Nacimiento: DD-MM-AAAA ")
fechaNacimiento = fechaNacimiento.replace('-','')

diaNacimiento = int(fechaNacimiento[0:2])
mesNacimiento = int(fechaNacimiento[2:4])
anoNacimiento = int(fechaNacimiento[4:8])
objetoNacimiento = datetime(year=anoNacimiento,month=mesNacimiento,day=diaNacimiento)
delta_18 = relativedelta(years=18)
deltaMayoria = fechaActual - delta_18
deltaNacimiento = fechaActual - objetoNacimiento

delta_relativo6 = relativedelta(years=6)
delta_relativo15 = relativedelta(years=15)

delta_6a = fechaActual - delta_relativo6
delta_15a = fechaActual - delta_relativo15

def ClasificacionTramo(fechaNac):
    #tramo 1
    if fechaNac > delta_6a:
        print(f"Tramo 1 - Menores de 6 años - Precio Entrada: $0")
    #Tramo 2
    elif fechaNac > delta_15a and fechaNac < delta_6a:
        print(f"Tramo 2 - Entre 6 y 14 años - Precio Entrada: $2000")
    #Tramo 3
    elif fechaNac > deltaMayoria and fechaNac < delta_15a:
        print(f"Tramo 3 - Entre 15 y 17 años - Precio Entrada: $4000")
    #Tramo 4
    elif fechaNac <= deltaMayoria:
        print(f"Tramo 4 - Mayores de 18 - Precio Entrada: $5000")

ClasificacionTramo(objetoNacimiento)
```

## 9° Ejercicio (0,5 puntos): 

Desarrolle un programa que tome un string como input y entregue el numero de apariciones de cada letra usando un loop.

```python
cadena = input("Ingrese una palabra: ").lower()

def ContadorCaracteres(cadena):
    numero_apariciones = {}
    print(f"---> {cadena}\n\nNúmero de apariciones por carácter:\n")
    for caracter in cadena:
        if caracter != " ":
            numero_apariciones[caracter] = numero_apariciones.get(caracter,0) + 1
    for caracter, numero in numero_apariciones.items():
        print(f"\t>>> {caracter} ---> {numero} veces")
    

ContadorCaracteres(cadena)
```

## 10° Ejercicio (1 puntos): 
Escriba una función que tome una lista de fechas en str como entradas en el formato "YYYY-MM-DD" y devuelva la menor y mayor fecha.

```python
from random import randint
from datetime import datetime

def GeneradorFechas(n):
    listaFechasAleatorias = []
    for i in range(n):
        fechaAleatoria = datetime(year=randint(1970,2023),month=randint(1,12),day=randint(1,28)) # <--- Variable 'month' sesgada a 28 días por mes
        fechaAleatoria = fechaAleatoria.strftime("%Y-%m-%d")
        listaFechasAleatorias.append(fechaAleatoria)
    return listaFechasAleatorias

def BuscadorMayorMenor(lista):
    lista.sort()
    fechaMenor = lista[0]
    fechaMayor = lista[len(lista)-1]
    print(f"Lista Input de Fechas Aleatorias:\n{lista}\n")
    print(f"Fecha más antigua ---> {fechaMenor}")
    print(f"Fecha más reciente ---> {fechaMayor}")

listaFechas = GeneradorFechas(randint(2,10))
BuscadorMayorMenor(listaFechas)
```
